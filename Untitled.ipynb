```python
import pandas as pd
url = 'https://raw.githubusercontent.com/LambdaSchool/data-science-practice-datasets/main/unit_1/Concrete/Concrete_Data.csv'
concrete = pd.read_csv(url)
concrete.head()
```

```python
# import the predictor class
from sklearn.linear_model import LinearRegression

# instantiate the class (with default parameters)
model = LinearRegression()

# assign the data to the x and y variables (easier to see what we are fitting)

x = concrete['cement']
# add a new axis to create a column vector
# (scikit-learn expects the data to be in this shape)
X = x[:, np.newaxis]

y = concrete['strength']

# fit the model
model.fit(X, y)

# slope (also called the model coefficient)
print('The slope: ', model.coef_)

# intercept
print('The intercept: ', model.intercept_)

# in equation form
print('Best-fit line: y = {:.2f}x + {:.2f}'.format(model.coef_[0], model.intercept_))
```

```python
import matplotlib.pyplot as plt

# plot the data
plt.scatter(x=x,y=y,alpha=0.5)

# plot the best-fit line
x_bestfit = np.linspace(100, 600, 10)
y_bestfit = 0.08 * x_bestfit + 13.44
plt.plot(x_bestfit, y_bestfit, color='red')

plt.xlabel('Cement content (kg/cubic meter)')
plt.ylabel('Compressive strength')
plt.title('Concrete Properties')
```

```python
# predict strength from a chosen cement content of 600
# y = 0.08x + 13.44 = 0.08 * 600 + 13.44 = 61

# calculate residual for one of the data points
# y = 0.08x + 13.44 = 0.08 * 198.6 + 13.44 = 29.3
# observed strength is 44.3, so residual is 44.30 - 29.30 = 15
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

crashes = sns.load_dataset('car_crashes')
mycols = ['total', 'speeding', 'alcohol']
crashes = crashes[mycols]
crashes.head()
```

```python
sns.lmplot(data=crashes,x='alcohol',y='total',ci=None)
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
import statsmodels.api as sm

Y = crashes['total']
X = crashes['alcohol']

# create the model with the X, Y data
X = sm.add_constant(X)
model = sm.OLS(Y,X)
# fit the model
results = model.fit()
# look at the results that include the t-statistic
print(results.t_test([1,0]))

# t-value : t = x bar - mu sub 0 / SE
# bar x = sample mean, mu sub 0 = mean under the null hypothesis, SE = standard error

# hypothesis test on slope parameter, we adjust t-statistic
# t = b sub 1 - beta sub 1 / SE
# b sub 1 = slope in the sample, beta sub 1 = slope under the null hyposthesis, SE = standard error

# in the example above, t = 5.8578 - 0 / 0.921 = 6.36
```

```python
# create a scatter plot with the confidence interval
sns.lmplot(data=crashes,x='alcohol',y='total',ci=95)
plt.show()
```
